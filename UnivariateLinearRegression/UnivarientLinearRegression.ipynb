```python

def plot_line_IS(intercept,slp,x_train,y_train):
    plt.scatter(x=x_train,y=y_train)
    plt.xlabel("x data -->")
    plt.ylabel("y data -->")
    plt.axline((0,intercept),slope=slp,color='r')
```

```python
plot_line_IS(4.93,.9240,x_train,y_train)
```

```python
def Model(x_train,intercept,slope):
    return ((slope*x_train)+intercept)
```

```python
Model(x_train,0,1)
```

```python
def cost_functionJ(b,w,x_train,y_train):
    model = Model(x_train,intercept=b,slope=w)
    cost_function = ((1/(2*model.size)) * np.sum((model-y_train)**2))
    return (model,cost_function)
```

```python
w = []
costs = []
b = []
for i in np.arange(-100,100,5):
    for j in np.arange(-90,90,5):
        b.append(i)
        w.append(j)
        costs.append(cost_functionJ(0,j,x_train,y_train)[1])
```

```python
plt.plot(w,costs)
plt.title("Cost function when W changes")
plt.xlabel("w value --->")
plt.ylabel("Cost Function J --->")
```

```python
b_test = []
cost_test = []
for i in range(-10,10,1):
    b_test.append(i)
    cost_test.append(cost_functionJ(i,1,x_train,y_train)[1])
    
plt.plot(b_test,cost_test)
plt.title("Cost function when b changes")
plt.xlabel("b value --->")
plt.ylabel("Cost Function J --->")
```

```python
fig = plt.figure()
ax = fig.add_subplot(111,projection='3d')
ax.scatter(np.array(b),np.array(w),np.array(costs),c='r',marker='o')
plt.show()
```
